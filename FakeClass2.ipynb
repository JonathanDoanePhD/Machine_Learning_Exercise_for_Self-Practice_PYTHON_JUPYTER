```python
import pandas as pd

df = pd.read_csv(r'C:\Users\kitsu\Documents\TempDocuments\MachineLearningFakeClass.csv')

#Let's count how many HW and Q each student missed
HW_miss = []
Q_miss = []

for s in range(len(df.index)):
    HW_miss.append(df.iloc[s,1:4].isnull().sum())
    Q_miss.append(df.iloc[s,4:7].isnull().sum())

df["HW_miss"] = HW_miss
df["Q_miss"] = Q_miss

#Failure to submit an assignment results in a "0" score
df.fillna(0, inplace=True)

#Now, let's tally up their scores
HW_tot = []
Q_tot = []

for s in range(len(df.index)):
    HW_tot.append(df.iloc[s,1:4].sum())
    Q_tot.append(df.iloc[s,4:7].sum())

df["HW_tot"] = HW_tot
df["Q_tot"] = Q_tot

#Let's make our target "Exam" column displayed last
df = df[ [ col for col in df.columns if col != 'Exam' ] + ['Exam']]

#Let's display our table as integers, rather than floats
pd.options.display.float_format = '{:,.0f}'.format

print(df)
```

```python
print("Which single feature is most indicative of Exam Score?", "\n")

from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error

# We aim to predict "y"
y = df.Exam

for feat in ["HW1", "HW2", "HW3", "Q1", "Q2", "Q3", "HW_miss", "Q_miss", "HW_tot", "Q_tot"]:
    X = df[[feat]]
    t_X, v_X, t_y, v_y = train_test_split(X, y, random_state = 0)
    print(feat, "\n")
    for max_leaf_nodes in range(2,11):
        model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
        model.fit(t_X, t_y)
        preds_val = model.predict(v_X)
        print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, mean_absolute_error(v_y, preds_val)))
    print('\n')
```

```python
print("My intuition:", "\n")
print(" * HW_miss and Q_miss lack variety, so the models based off of those columns lack discriminants", "\n")
print(" * HW3 is the worst, likely caused by the (HW3,Exam) pairs ranging from (Good,Bad), (Good,Good), (Bad, Good), (Bad,Bad)!", "\n")
print(" * Q_tot is the most indicative; close behind are Q_2 and even HW_1, HW_2", "\n")
print(" * HW1, HW2 are fairly good", "\n")
print(" * It seems my (fake) homework wasn't synchronized with the Exam material very well...")
```

```python
print("I wonder what would happen if I used the best few indicators for my next model?", "\n")

X = df[["HW1", "HW2", "Q_tot", "Q2"]]

t_X, v_X, t_y, v_y = train_test_split(X, y, random_state = 0)

print("HW1, HW2, Q2, Q_tot", "\n")
for max_leaf_nodes in range(2,11):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(t_X, t_y)
    preds_val = model.predict(v_X)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, mean_absolute_error(v_y, preds_val)))
```

```python
print("Interesting...  Well, let's try a few more models before concluding.", "\n")

from itertools import combinations

possible_feats = ["HW1", "HW2", "Q_tot", "Q2"]

for i in [2,3,4]:
    for feats in combinations(possible_feats, i):  # 2 for pairs, 3 for triplets, etc
        print(feats, "\n")
        X = df[list(feats)]
        t_X, v_X, t_y, v_y = train_test_split(X, y, random_state = 0)
        for max_leaf_nodes in range(2,11):
            model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
            model.fit(t_X, t_y)
            preds_val = model.predict(v_X)
            print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, mean_absolute_error(v_y, preds_val)))
        print("\n")
```

```python
print("I believe Q_tot is the singlemost indicative column for Exam score, and altogether Q_tot, Q2, and HW1 are fairly indicative as well.", "\n")

print("Let's end with a (hopefully very) nice model.")

feats = ["HW1", "Q2", "Q_tot"]

X = df[feats]

model = DecisionTreeRegressor(max_leaf_nodes=4, random_state=0)
model.fit(X, y)
preds_val = model.predict(X)

myX = pd.DataFrame(X)

myX["Exam"] = y
myX["Pred"] = preds_val
print("\n")
print(myX)
myX.drop(columns=["Exam", "Pred"], inplace=True)

print("\n")
print("MAE:", mean_absolute_error(y, preds_val))
```
